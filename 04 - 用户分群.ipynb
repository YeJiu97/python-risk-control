# 辛普森悖论

定义

辛普森悖论（Simpson's Paradox）是统计学中的一种现象，它表明在数据的分组分析和整体分析中，可能会出现相互矛盾的结论。这种悖论发生的原因在于，当不同子组的数据合并后，整体的趋势可能与每个子组的趋势相反。

简单来说，辛普森悖论揭示了数据在不同层级下的解读可能会有所不同。当不加区分地将不同组别的数据合并分析时，隐藏的因素（如混杂变量）可能会导致得出错误的结论。只有在对这些组别单独分析的情况下，才能揭示出更准确的趋势。

举例：

假设一个医院在不同时间段对两组病人进行了治疗。在每个时间段内，某个特定的治疗方法似乎对每个病人群体都有效，但当将所有时间段的数据合并起来分析时，可能会得出这个治疗方法无效甚至有害的结论。这种看似矛盾的现象就是辛普森悖论的体现。

辛普森悖论告诉我们，数据分析时需要特别小心如何对数据进行分组和汇总，避免得出误导性的结论。

为了演示辛普森悖论，我们可以使用一个具体的例子。假设我们有两家医院，A 和 B，各自在两个季度内对病人的治愈情况进行了记录。数据如下：

| 医院   | 季度 | 治愈人数 | 总人数 | 治愈率（治愈人数/总人数） |
|------|------|--------|------|------------------|
| A    | Q1   | 30     | 50   | 60%              |
| A    | Q2   | 10     | 50   | 20%              |
| B    | Q1   | 15     | 30   | 50%              |
| B    | Q2   | 35     | 70   | 50%              |

单独看每个季度：

- 在Q1：医院A的治愈率为 60%（30/50），医院B的治愈率为 50%（15/30），因此看起来医院A表现得更好。
- 在Q2：医院A的治愈率为 20%（10/50），医院B的治愈率为 50%（35/70），因此医院B表现得更好。

如果将数据合并：

合并医院A和B两个季度的数据如下：

| 医院   | 总治愈人数 | 总人数 | 总治愈率 |
|------|--------|------|------|
| A    | 40     | 100  | 40%  |
| B    | 50     | 100  | 50%  |

在整体合并的情况下，医院B的治愈率为 50%，而医院A的治愈率为 40%。这样看起来，医院B的总体表现似乎更好。

辛普森悖论的出现：

- 如果单独分析每个季度，Q1中医院A表现得比医院B好，而Q2中医院B表现得比医院A好。
- 但在整体合并后，医院B的总体治愈率高于医院A。

这是因为在不同的季度中，每家医院的治疗人数和成功率不均衡。医院A在第一季度治疗了较多病人，治愈率较高，但第二季度人数相对较少，治愈率较低。而医院B的两季度治愈率都保持了一致。因此，合并后的数据掩盖了医院A在某些情况下的相对优势，从而出现了辛普森悖论的效果。

这个例子说明了在数据分析中，必须谨慎处理分组数据，特别是在处理潜在混杂变量或分层分析时。

# 决策树

## 决策树

决策树是一种基于树状结构的机器学习模型，常用于分类和回归任务。它通过一系列的规则将数据分割成不同的组，以实现对目标变量的预测。每个分支代表一个决策过程中的条件，每个叶节点（终点）表示预测结果或决策结果。


决策树的基本组成部分：
1. 节点：
   - 根节点：决策树的起点，包含整个数据集，表示模型的第一个决策条件。
   - 内部节点：每个内部节点对应于一个决策，通常基于某个特征值或条件进行数据的分割。
   - 叶节点：叶节点是决策树的终点，表示最终的分类或回归预测结果。
2. 分支：从一个节点到另一个节点的路径，表示某个特征的取值范围或某个条件的选择。

决策树的工作原理：
决策树通过递归地分割数据集，逐步找到能最大程度上减少不纯度的特征进行分割。常用的分割标准有：
- 信息增益（Information Gain）：基于信息论中的熵来衡量分裂后的纯度变化。信息增益越大，分裂后的纯度越高。
- 基尼指数（Gini Index）：用来衡量数据的混杂程度，基尼指数越小表示越纯。
- 方差减少（Variance Reduction）：主要用于回归问题，用于衡量分裂后方差的减少程度。

决策树的优缺点

优点：
1. 易于理解和解释：决策树模型类似于人类的思维过程，决策路径可以很容易地解释给非技术人员。
2. 不需要数据预处理：它不要求数据进行标准化或归一化，甚至可以处理缺失值。
3. 能够处理数值型和类别型数据：决策树可以同时处理离散和连续的特征。
4. 基于特征的非线性关系：决策树擅长处理非线性关系，特别是在特征之间存在复杂交互时。

缺点：
1. 容易过拟合：如果决策树的深度过大，它可能会对训练数据产生过拟合的现象。为了解决这个问题，可以通过剪枝（pruning）或者限制树的深度等方法进行控制。
2. 对数据变化敏感：小的噪声或数据变化可能会导致完全不同的决策树结构。
3. 偏向于重要特征：决策树倾向于优先选择具有较多取值范围的特征，可能导致一些重要但较少变化的特征被忽略。

决策树的应用：
- 分类任务：如垃圾邮件检测、客户细分、肿瘤诊断等。
- 回归任务：如房价预测、市场预测等。
- 特征重要性分析：决策树可以用来评估不同特征的重要性，帮助解释模型的预测。

决策树的示例

假设我们有一个学生数据集，目标是根据特征预测学生是否会通过考试，数据包括：
- 特征1：学习时间
- 特征2：是否参加辅导班

决策树可能的分割过程如下：
1. 如果学习时间 > 2 小时，则继续分割。
2. 如果学习时间 ≤ 2 小时，且参加辅导班，则预测通过考试。
3. 如果学习时间 ≤ 2 小时，且不参加辅导班，则预测未通过考试。

通过这棵决策树，我们可以明确地知道决策流程，并且可以很容易地解释每个分裂条件。

````python
df = pd.read_csv("CreditScoring.csv")  # 导入对应的数据集
````

````python
# 可以发现所有的列都是大写的，虽然都是统一格式，但是这里将其改写为小写：
df.columns = df.columns.str.lower()
df.head()
````

````python
# 这里开始进行转换
status_values = {
    1: 'ok',  # 1表示没有问题
    2: 'default',  # 2表示违约
    0: 'unk'  # 0表示unknown，也就是值缺失
}
````

````python
# 然后开始进行映射
df.status = df.status.map(status_values)
````

````python
# 对其他的变量也执行相同的操作
# home变量
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}
df.home = df.home.map(home_values)

# 婚恋变量
marital_values = {
    1: 'single',
    2: 'married',
    3: 'widow',
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}
df.marital = df.marital.map(marital_values)

# 过往记录
records_values = {
    1: 'no',
    2: 'yes',
    0: 'unk'
}
df.records = df.records.map(records_values)

# 工作情况
job_values = {
    1: 'fixed',
    2: 'partime',
    3: 'freelance',
    4: 'others',
    0: 'unk'
}
df.job = df.job.map(job_values)
````

````python
df.describe().round()  # 描述性统计
````

````python
# 缺失值替换
for c in ['income', 'assets', 'debt']:
    df[c] = df[c].replace(to_replace = 99999999, value = np.nan)  # to_replace为需要被替换的值，value为用来进行替换的值
````

````python
df.status.value_counts()
````

````python
df[df.status == 'unk']
````

````python
df = df[df.status != 'unk']  # 这里是创建了一个新的DataFrame，但是不包含unk的status数据行
````

````python
from sklearn.model_selection import train_test_split
 
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=11)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=11)
````

````python
y_train = (df_train.status == 'default').values
y_val = (df_val.status == 'default').values
````

然后是从原先的数据集中删除：

````python
del df_train['status']
del df_val['status']
````

````python
df_train = df_train.fillna(0)
df_val = df_val.fillna(0)
````

````python
# 为了使用分类变量，还需要进行独热编码：
dict_train = df_train.to_dict(orient = 'records')
dict_val = df_val.to_dict(orient = 'records')
````

````python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse = False)

X_train = dv.fit_transform(dict_train)
X_val = dv.transform(dict_val)
````

````python
from sklearn.tree import DecisionTreeClassifier  # 导入对饮的包

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
````

````python
from sklearn.metrics import roc_auc_score

y_pred = dt.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, y_pred)
````

````python
# 但是我们需要注意的是：在训练集上的100%完美预测既有可能是因为模型的性能特别的好，也有可能是因为模型在训练集上出现了过拟合的情况。
# 使用验证集来进行测试
y_pred = dt.predict_proba(X_val)[:, 1]
roc_auc_score(y_val, y_pred)
````

可以发现，在验证集上，只有65的准确率。

可以发现，原先的模型是一个典型的过拟合的情况，这是决策树本身的原理决定的：



决策树可以不断地增加节点和深度，从而为每一个样本都给出一个最终节点，从而完美完美的分类。



````python
# 默认情况之下，树的深度参数为None
dt = DecisionTreeClassifier(max_depth = 2)
dt.fit(X_train, y_train)
````

````python
# 可以使用tree包中的export_text函数来进行可视化：

from sklearn.tree import export_text

tree_text = export_text(dt, feature_names=dv.feature_names_)
print(tree_text)
````

````python
from sklearn import tree

plt.figure(figsize=(10, 6))
tree.plot_tree(dt, feature_names=dv.feature_names_, filled=True, rounded=True, fontsize=10)
plt.show()
````

````python
# 接着看一下这个决策树的性能如何：

y_pred = dt.predict_proba(X_train)[:, 1]
auc = roc_auc_score(y_train, y_pred)
print('train auc', auc)

y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print("validation auc", auc)
````

从结果来看，可以发现决策树在训练集上的准确度下降了，但是在验证集上的准确性上升了。

但是这个树太简单了，所以需要尝试不同的参数并且查看哪些参数的效果最佳，能够带来最好的AUC。

````python
# 尝试不同的参数，并且计算对应的auc分数
for depth in [1, 2, 3, 4, 5, 6, 10, 15, 20, None]:
    dt = DecisionTreeClassifier(max_depth = depth)
    dt.fit(X_train, y_train)
    y_pred = dt.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    print('%4s -> %.3f' % (depth, auc))
````

可以发现当max_depth的值为5的情况之下是最好的。

但是我们需要采用4，5，6三个max_depth，然后调试一下min_samples_leaf的参数：

````python
for m in [4, 5, 6]:
    
    print('depth: %s' % m)

    for s in [1, 5, 10, 15, 20, 50, 100, 200]:
        dt = DecisionTreeClassifier(max_depth = m, min_samples_leaf = s)
        dt.fit(X_train, y_train)
        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
        print('%4s -> %.3f' % (s, auc))
````

结果最好的是depth=6和min_samples_leaf=15。

````python
# 使用这个最佳的参数来训练最后的树
dt = DecisionTreeClassifier(max_depth = 6, min_samples_leaf=15)
dt.fit(X_train, y_train)
````

````python
plt.figure(figsize=(20, 12))
tree.plot_tree(dt, feature_names=dv.feature_names_, filled=True, rounded=True, fontsize=10)
plt.show()
````

## 信息增益

信息增益（Information Gain） 是决策树构建过程中常用的一种度量，用来衡量某个特征在数据集划分时带来的“纯度”提升。其目的是帮助选择最优的特征来进行数据分割，以便提高分类效果。

信息增益基于信息论中的“熵”概念。熵用于衡量数据集的混乱程度，或者说系统中不确定性的大小。熵值越高，表示数据越混乱；熵值越低，表示数据越纯净。

信息增益的计算过程：
1. 计算数据集的总熵：
   首先，计算整个数据集的熵，表示当前数据的混乱程度。公式为：

   $
   H(D) = - \sum_{i=1}^{n} p_i \log_2(p_i)
   $

   其中 $p_i$ 表示分类为 $i$ 类的样本占整个数据集的比例，$n$ 是类别的数量。

2. 根据特征分割数据集：
   接下来，选择一个特征，并根据该特征的不同取值将数据集划分为多个子集。

3. 计算分割后的熵：
   对每个子集计算熵值，然后根据每个子集的样本比例计算整个数据集的条件熵。条件熵的公式为：

   $
   H(D|A) = \sum_{v \in V(A)} \frac{|D_v|}{|D|} H(D_v)
   $

   其中 $A$ 是当前特征，$V(A)$ 是特征的取值集合，$D_v$ 是特征 $A$ 取值为 $v$ 时对应的子集，$|D_v|$ 是这个子集的样本数，$|D|$ 是数据集的总样本数。

4. 计算信息增益：
   信息增益就是原数据集的熵与根据特征分割后的条件熵之间的差值，公式为：

   $
   IG(D, A) = H(D) - H(D|A)
   $

   信息增益表示通过选择特征 $A$ 进行数据分割后，数据混乱度降低了多少。信息增益越大，说明这个特征对数据的分割效果越好。

举例说明：

假设我们有一个简单的二分类数据集，用于预测是否会下雨。特征为“是否有云”和“湿度”。数据如下：

| 是否有云 | 湿度   | 是否下雨 |
|--------|------|--------|
| 是      | 高    | 是      |
| 是      | 高    | 是      |
| 否      | 高    | 否      |
| 否      | 低    | 否      |
| 是      | 低    | 是      |

第一步：计算总体熵

假设我们有5个样本，其中3个是“下雨”，2个是“不下雨”，总体熵 $H(D)$ 为：

$
H(D) = - \left( \frac{3}{5} \log_2 \frac{3}{5} + \frac{2}{5} \log_2 \frac{2}{5} \right) = 0.971
$

第二步：根据特征“是否有云”进行分割

假设我们用“是否有云”这个特征来分割数据：
- 如果“有云”，则数据为{是, 是, 是}，熵为 0。
- 如果“无云”，则数据为{否, 否}，熵为 0。

条件熵 $H(D|\text{是否有云})$ 为：

$
H(D|\text{是否有云}) = \frac{3}{5} \times 0 + \frac{2}{5} \times 0 = 0
$

第三步：计算信息增益

信息增益 $IG(D, \text{是否有云})$ 为：

$
IG(D, \text{是否有云}) = H(D) - H(D|\text{是否有云}) = 0.971 - 0 = 0.971
$

这个结果表明，“是否有云”这个特征提供了很高的信息增益，因此是一个好的分裂特征。

信息增益的作用：

在决策树算法（如 ID3 算法）中，信息增益用于选择最佳的特征来分割数据集，帮助模型在树的每一层做出最优决策。

## 基尼指数

基尼指数（Gini Index） 是决策树（尤其是 CART算法）中用于衡量数据集的不纯度或混乱程度的一种指标。它表示在一个数据集中，随机选取两个样本被错误分类的概率。基尼指数越小，数据集的纯度越高，分类的准确性越高。因此，在构建决策树时，通常会选择基尼指数最小的特征进行数据分割。

基尼指数的定义：

对于一个类别型数据集，如果数据集中的样本可以被分为 $ n $ 个类别，基尼指数的公式如下：

$
Gini(D) = 1 - \sum_{i=1}^{n} p_i^2
$

其中：
- $ p_i $ 是数据集中第 $ i $ 类的样本所占的比例（概率）。
- $ n $ 是类别的数量。

基尼指数的取值范围在 [0, 1] 之间：
- 0 表示数据集非常纯净，所有样本属于同一个类别。
- 1 表示数据集混乱程度最高，所有类别的样本均匀分布。

基尼指数的工作原理：
在构建决策树时，基尼指数用于评估某个特征对数据集的分割效果。通过计算基于该特征进行分割后的基尼指数（即条件基尼指数），我们可以选择最优的特征来分割数据集。

基尼指数的计算步骤：
1. 计算整体数据集的基尼指数。
2. 根据某个特征将数据集进行分割，得到多个子集。
3. 计算每个子集的基尼指数，并根据子集的样本比例计算加权后的总基尼指数（即条件基尼指数）。
4. 选择基尼指数最小的分割，因为基尼指数越小，数据集越纯净。

基尼指数的示例：
假设我们有一个简单的二分类数据集，目标是预测是否会下雨。样本如下：

| 是否有云 | 湿度   | 是否下雨 |
|--------|------|--------|
| 是      | 高    | 是      |
| 是      | 高    | 是      |
| 否      | 高    | 否      |
| 否      | 低    | 否      |
| 是      | 低    | 是      |

我们来计算基尼指数，假设我们使用“是否有云”这个特征来分割数据：

1. 计算整体数据集的基尼指数：
   在这个数据集中，3个样本是“下雨”，2个样本是“不下雨”。总体基尼指数为：

   $
   Gini(D) = 1 - \left(\frac{3}{5}\right)^2 - \left(\frac{2}{5}\right)^2 = 1 - 0.36 - 0.16 = 0.48
   $

2. 根据“是否有云”进行分割：
   - 有云时，数据为{是, 是, 是}，所有样本都属于同一个类别，因此基尼指数为 0。
   - 无云时，数据为{否, 否}，基尼指数也是 0，因为所有样本也属于同一个类别。

3. 计算分割后的加权基尼指数：
   分割后的加权基尼指数为：

   $
   Gini(\text{是否有云}) = \frac{3}{5} \times 0 + \frac{2}{5} \times 0 = 0
   $

4. 结论：
   通过“是否有云”这个特征进行分割，得到了纯度很高的数据集（即基尼指数为0），因此这个特征是一个很好的分割特征。

基尼指数与信息增益的区别：
- 信息增益基于“熵”的概念，用于衡量分裂后不确定性（混乱度）的减少。
- 基尼指数用于直接衡量分类的不纯度（错误分类的概率），计算上通常比信息增益更简单和快速。

在实际应用中，CART算法通常使用基尼指数作为决策树分裂标准，而ID3算法则使用信息增益。

总结：
基尼指数作为决策树中的一种重要指标，帮助选择最佳的分裂特征。基尼指数越小，表示分裂后的数据更加纯净，分类效果更好。因此，决策树构建时倾向于选择基尼指数较小的分裂点。

## CART树

CART树（Classification and Regression Tree，分类与回归树）是一种常用的决策树算法，能够同时处理分类任务和回归任务。CART树通过递归地将数据集划分为子集，以构建二叉树结构，每个节点基于特征的某个条件进行分裂，最终叶节点给出预测结果。

**CART树的两种类型：**
1. 分类树（Classification Tree）：
   - 用于解决分类问题，目标是根据特征将样本划分到不同的类别中。
   - 典型的输出是类别标签。
   - 分类树的常用分裂标准是基尼指数，用于衡量数据集的不纯度。

2. 回归树（Regression Tree）：
   - 用于解决回归问题，目标是预测一个连续的数值。
   - 典型的输出是一个数值（如房价预测）。
   - 回归树的分裂标准通常是方差，用于衡量数据的分散程度。

**CART树的工作原理：**

CART树是一种二叉树模型，这意味着每次分裂时数据集被分成两部分。其构建过程如下：
1. 选择最优的分裂特征和分裂点：对于每一个特征，CART树会计算不同的分裂点，使用基尼指数（分类树）或方差（回归树）来选择最优的分裂方式，使得分裂后的数据子集更加纯净或差异性更小。
2. 递归地分割数据：一旦确定了最佳分裂点，CART树将数据分成两个子集，并在每个子集上继续重复上述过程，直到满足终止条件。
3. 生成叶节点：当到达某些终止条件（如数据集已经纯净、样本数太少、树的最大深度达到等）时，停止分裂，并将当前节点作为叶节点。对于分类任务，叶节点包含分类结果；对于回归任务，叶节点包含预测的数值。

**树的剪枝：**
CART树容易产生过拟合，因为树可能会过度分割数据，导致其对训练数据拟合过好，从而在测试数据上表现不佳。为了避免过拟合，通常会使用剪枝（Pruning）技术来简化树结构：
- 预剪枝：在构建树的过程中提前停止分裂，比如设定最大树深度、最小样本数等条件，防止树过深。
- 后剪枝：构建完决策树后，通过评估叶节点的表现，去除不必要的分支，使树变得更加简洁。

**CART树的优点：**
1. 易于解释：CART树的模型结构类似于一系列条件判断，非常直观，便于理解和解释。
2. 能够处理连续型和离散型数据：CART树可以同时处理类别型特征（如性别）和数值型特征（如年龄）。
3. 非参数模型：不需要对数据分布做出任何假设，能够处理非线性关系。
4. 适应性强：CART树可以用于分类和回归问题，应用范围广泛。

**CART树的缺点：**
1. 容易过拟合：如果不使用剪枝技术，决策树可能会对训练数据产生过拟合，导致泛化能力差。
2. 对数据敏感：小的噪声或数据变化可能会导致树的结构发生较大的变化。
3. 局部最优：由于CART树是贪心算法，在每个分裂步骤中只寻找局部最优的分割点，可能无法找到全局最优解。

**举例说明 CART 分类树：**
假设我们有一个数据集，用于预测某人是否会购买某产品，特征包括“年龄”和“收入”，目标是分类“是否购买”：
- 如果“年龄 ≤ 30”，继续判断收入。
- 如果“收入 > 50k”，预测购买；否则，预测不购买。
- 如果“年龄 > 30”，直接预测不购买。

这就是一个典型的 CART 分类树，通过逐步分割特征空间，将数据划分到不同的类别中。

**CART树与其他决策树算法的区别：**
1. ID3 和 C4.5 决策树：
   - 这两种算法主要用于分类问题，ID3 使用信息增益作为分裂标准，而 C4.5 使用信息增益率。
   - 与 CART 树不同，它们生成的是多叉树，而 CART 树是二叉树。
2. CART 树：
   - CART 同时可以处理分类和回归任务。
   - 使用基尼指数（分类）或方差（回归）来选择最优分裂。
   - CART树生成的始终是二叉树。

**总结：**
CART树是一种灵活、直观的机器学习模型，广泛用于分类和回归任务。它在实践中具有良好的性能，并且易于解释，但也需要通过剪枝等技术来避免过拟合。

## GMM

GMM（Gaussian Mixture Model，高斯混合模型）是一种统计模型，用于表示一个包含多个高斯分布的复合概率分布。GMM广泛应用于聚类分析、密度估计、图像处理等领域。

GMM的主要特点：

1. **混合分布**：GMM假设数据点是由多个高斯分布生成的，每个高斯分布代表数据的一个簇或类别。

2. **参数化模型**：每个高斯分布由其均值（mean）、方差（variance）和权重（weight）参数定义。模型的整体参数由所有高斯分布的参数组成。

3. **期望最大化（EM）算法**：GMM通常使用EM算法进行参数估计。该算法包括两个步骤：
   - **期望步骤（E步）**：计算每个数据点属于各个高斯分布的概率。
   - **最大化步骤（M步）**：根据E步的结果更新高斯分布的参数。

4. **应用广泛**：GMM可用于聚类、图像分割、异常检测等任务，能够处理不规则形状的数据分布。

示例应用：

- 在图像处理中，GMM可用于对图像进行分割，通过识别不同颜色或亮度的区域。
- 在市场细分中，GMM可以帮助识别客户群体中的不同类型或偏好。

GMM因其灵活性和强大的建模能力，成为许多机器学习和统计分析中的重要工具。

高斯分布（Gaussian distribution），又称正态分布（Normal distribution），是一种重要的概率分布，广泛应用于统计学、自然科学和社会科学等领域。其概率密度函数（PDF）具有特定的数学形式，并且其形状呈现出对称的钟形曲线。

**高斯分布的主要特点：**

1. **数学表达**：
   高斯分布的概率密度函数为：
   $
   f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
   $
   其中：
   - $ \mu $ 是均值，表示分布的中心位置。
   - $ \sigma $ 是标准差，表示分布的离散程度，标准差越大，分布越宽。

2. **对称性**：
   高斯分布是关于均值对称的，即在均值两侧的概率分布是相等的。

3. **68-95-99.7法则**：
   - 约68%的数据落在均值±1个标准差的范围内。
   - 约95%的数据落在均值±2个标准差的范围内。
   - 约99.7%的数据落在均值±3个标准差的范围内。

4. **中心极限定理**：
   中心极限定理指出，许多独立随机变量的和在适当的归一化后趋向于高斯分布，即使这些变量本身的分布不是正态的。这使得高斯分布在统计推断中极为重要。

5. **应用广泛**：
   高斯分布在各种领域中都有应用，包括：
   - 自然科学（如物理学、生物学）
   - 社会科学（如心理学、经济学）
   - 机器学习（如数据分析、模式识别）

由于其良好的数学性质和广泛的适用性，高斯分布被视为统计学和概率论中的基础分布之一。

## K均值聚类

K均值聚类是一种**无监督学习算法**，用于将数据分成**K个簇（clusters）**。它的目标是通过**最小化簇内样本之间的距离**，使得同一簇中的数据点尽可能相似，而不同簇之间的差异尽可能大。算法的基本步骤如下：

1. **选择K个簇中心**：随机选择K个初始中心点（即质心，centroids）。
2. **分配数据点到最近的簇**：根据每个数据点与各个中心点的距离（通常使用欧氏距离），将每个点分配到最近的簇中心。
3. **更新簇中心**：重新计算每个簇的中心点，新的中心是该簇中所有数据点的**平均值**。
4. **重复步骤2和3**：不断更新数据点的分配和簇中心，直到簇中心不再发生变化或变化非常小。

**优点：**
- 简单且易于理解和实现。
- 在处理大型数据集时较为高效。

**缺点：**
- 需要事先指定簇的数量K，这对实际问题可能不直观。
- 对初始点的选择敏感，可能会陷入局部最优解。
- 只适合形状较为规则的簇，不适用于复杂形状的簇。

总结来说，K均值聚类通过反复调整簇中心和分配数据点来找到最优的簇划分，使得簇内的相似性最大化。

````python
#K-均值聚类
from sklearn import datasets
from sklearn.cluster import KMeans

iris_df = datasets.load_iris()

model = KMeans(n_clusters=3)

model.fit(iris_df.data)

predicted_label = model.predict([[7.2, 3.5, 0.8, 1.6]])

all_predictions = model.predict(iris_df.data)

print(predicted_label)
print(all_predictions)
````

## 层次聚类

**层次聚类（Hierarchical Clustering）**是一种构建嵌套聚类结构的无监督学习算法，它根据数据点的相似性，逐步形成一个树状结构（称为**聚类树**或**树状图**，dendrogram）。该算法分为两类：

1. **凝聚层次聚类（Agglomerative Hierarchical Clustering）**：从**每个数据点作为一个单独的簇**开始，逐步将最相似的簇合并，直到所有数据点合并为一个簇。
2. **分裂层次聚类（Divisive Hierarchical Clustering）**：从**所有数据点作为一个簇**开始，逐步将簇拆分，直到每个数据点成为一个单独的簇。

**凝聚层次聚类的步骤：**
1. **计算距离**：首先计算所有数据点之间的相似度（通常使用欧氏距离、曼哈顿距离等）。
2. **合并最相似的两个簇**：找到距离最近的两个簇，并将它们合并为一个簇。
3. **更新距离矩阵**：根据新簇与其他簇的距离重新计算距离矩阵。
4. **重复合并**：不断重复合并操作，直到所有数据点合并成一个簇，或者达到预定的聚类数量。

**分裂层次聚类的步骤：**
1. **将所有数据点看作一个簇**。
2. **递归分裂簇**：根据簇中数据点的相似性，将簇逐步划分为较小的簇。
3. **重复分裂**：不断分裂，直到每个数据点成为一个独立的簇，或者达到预定的聚类数量。

**聚类树（树状图）的解读：**

层次聚类的结果可以通过 **树状图（dendrogram）** 来表示，它展示了数据点的合并过程。树状图的垂直轴代表聚类的距离，水平轴代表数据点。通过在树状图上选择一个切割点，可以选择不同层次的聚类结果（即生成不同数量的簇）。

**优点**：
- 不需要预先指定簇的数量，用户可以根据树状图选择合适的聚类层次。
- 对于形状复杂的簇也可以有效聚类，不像K均值那样依赖于球形簇。

**缺点**：
- 计算复杂度高，尤其是当数据集很大时，计算距离矩阵和更新距离的过程会非常耗时。
- 一旦合并或分裂，不能回退调整，因此对噪声或异常值敏感。

总结来说，层次聚类通过构建聚类树，逐步形成从个体到整体的聚类结构，适合探索性数据分析和理解数据的内在结构。

````python
#t-SNE 聚类
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

iris_df = datasets.load_iris()

model = TSNE(learning_rate=100)

transformed = model.fit_transform(iris_df.data)

x_axis = transformed[:, 0]
y_axis = transformed[:, 1]

plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.show()
````

## DBSCAN聚类

````python
#DBSCAN 聚类的
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

iris = load_iris()

dbscan = DBSCAN()

dbscan.fit(iris.data)

pca = PCA(n_components=2).fit(iris.data)
pca_2d = pca.transform(iris.data)

for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')

plt.legend([c1, c2, c3], ['Cluster 1', 'Cluster 2', 'Noise'])
plt.title('DBSCAN finds 2 clusters and Noise')
plt.show()
````
