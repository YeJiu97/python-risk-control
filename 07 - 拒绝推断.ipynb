# 拒绝推断

拒绝推断（Rejection Inference）是一种用于改善模型预测准确性的方法，尤其是在信用评分和风险评估等领域。这种方法主要应用于以下情况：

1. **数据不平衡问题**：在信用评分中，通常只有部分申请人会被批准，而拒绝的申请人信息则较少。这导致模型在训练时无法获得完整的数据分布，可能会导致偏差。

2. **增强训练数据集**：拒绝推断的核心思想是通过将拒绝申请者的数据整合到模型中，以便更好地捕捉申请者的特征和行为，从而提高模型对未来申请者的预测能力。

**拒绝推断的步骤**

1. **数据收集**：收集被拒绝申请者的特征和相关信息。

2. **模型构建**：首先使用被批准申请者的数据来构建初步模型。

3. **推断拒绝者**：通过对拒绝申请者的特征进行建模，预测他们如果获得批准可能的结果。这样可以从中推断出拒绝者的潜在信用风险。

4. **合并数据**：将被拒绝者的预测结果与批准者的数据结合，形成一个更全面的训练集。

5. **重新建模**：使用合并后的数据集重新训练模型，以提高对申请者的整体预测能力。

**优势与挑战**

- **优势**：
  - 提高模型的准确性：通过引入拒绝者的信息，模型能够更全面地理解申请者的特征。
  - 更好地评估风险：通过包括拒绝者的特征，金融机构可以更好地评估信用风险。

- **挑战**：
  - 数据质量：拒绝者的信息可能不完整或不准确，可能影响模型的有效性。
  - 道德与法律问题：在某些情况下，使用拒绝者的数据可能会引发隐私和公平性的问题。

拒绝推断是一种强有力的工具，但在实施时需要谨慎，以确保数据的合理性和模型的公正性。

# Python例子

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# 创建模拟数据
data = {
    'age': [25, 45, 35, 50, 23, 43, 36, 30, 29, 41],
    'income': [50000, 60000, 58000, 80000, 40000, 75000, 62000, 52000, 54000, 78000],
    'credit_score': [700, 750, 720, 780, 680, 730, 710, 690, 670, 760],
    'approved': [1, 1, 1, 1, 0, 1, 1, 0, 0, 1]  # 1 表示批准，0 表示拒绝
}

df = pd.DataFrame(data)

# 将拒绝者的数据保留为另一个 DataFrame
rejected = df[df['approved'] == 0]

# 数据集划分
X = df.drop(columns='approved')
y = df['approved']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 构建初步模型
model = LogisticRegression()
model.fit(X_train, y_train)

# 计算预测概率
y_pred_prob = model.predict_proba(X_test)[:, 1]

# 计算 AUC
auc_score = roc_auc_score(y_test, y_pred_prob)
print(f'Initial AUC Score: {auc_score}')

# 拒绝推断
# 使用拒绝者的信息进行推断（在这里我们简单假设拒绝者的批准概率是 0.5）
rejected['predicted_prob'] = 0.5  # 这是一个假设值，实际情况中会根据特征进行计算

# 合并数据
rejected_approved = rejected.drop(columns='approved').reset_index(drop=True)
predicted_probs = pd.DataFrame({'predicted_prob': y_pred_prob}).reset_index(drop=True)

# 确保两个数据集的行数一致
combined = pd.concat([predicted_probs, rejected_approved], axis=0, ignore_index=True)

# 合并后的数据集
X_combined = combined
y_combined = np.concatenate([y_test.reset_index(drop=True), np.zeros(rejected.shape[0])])  # 拒绝者标签为0

# 检查并处理缺失值
if X_combined.isnull().values.any():
    X_combined = X_combined.fillna(X_combined.mean())  # 用均值填充缺失值

# 重新训练模型
model_combined = LogisticRegression()
model_combined.fit(X_combined, y_combined)

# 计算新的预测概率和 AUC
y_pred_prob_combined = model_combined.predict_proba(X_combined)[:, 1]
new_auc_score = roc_auc_score(y_combined, y_pred_prob_combined)
print(f'New AUC Score after Rejection Inference: {new_auc_score}')

# 可视化 ROC 曲线
fpr, tpr, thresholds = roc_curve(y_combined, y_pred_prob_combined)

plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, label='ROC Curve (AUC = %0.2f)' % new_auc_score)
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Receiver Operating Characteristic')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```
