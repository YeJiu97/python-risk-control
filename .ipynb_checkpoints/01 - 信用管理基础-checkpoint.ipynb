# 信贷与管理

信贷业务也被称为信贷资产或者贷款业务，是商业银行和互联网金融公司最重要的资产业务和主要盈利手段。

信贷管理主要包含了两个概念：

- $信用$：先买后付，通过信用值来预支付金钱以购买相应服务
- $管理$：通过用户信息和对用户信用度进行评估，并且根据信用情况来制定风险规避策略

所谓的风控：$对用户风险进行管理和规避的过程$。

信贷管理领域，一个非常重要的概念叫做：$风险数据分析$，它对于用户的信用风险进行管理与规避。

对于预测信用较差的人，贷款机构一般是不会向其放贷的，即便是放贷，也会为了抵御风险给予其较高的利息和较小的贷款金额。

对历史数据使用数学公式进行拟合学习，可以得到模型，但是使用模型输出的也只是未来事件的预期。

# 信贷基础指标

**年度百分率（APR）：复利计息，通常是按一年一年来计算的利率。**


年利率的计算公式可以是：
$ \text{年利率} = \left( \frac{\text{年度利息}}{\text{本金}} \right) \times 100\% $

这里的“年度利息”是指一年内获得的利息总额，“本金”是指最初借出或存入的金额。

需要注意的是，年利率可以是简单年利率（Simple Interest Rate）或复利年利率（Compound Interest Rate）：

1. **简单年利率**：如果利息不复利，即利息不会产生额外的利息，那么年利率就是简单年利率。
2. **复利年利率**：如果利息在一年内多次计算并加入本金中产生新的利息，那么年利率就是复利年利率。

复利年利率的计算更为复杂，因为它涉及到周期性的利息计算，公式如下：

$ A = P \left(1 + \frac{r}{n}\right)^{nt} $
其中：
- $ A $ 是账户的未来价值，包括利息。
- $ P $ 是本金（初始存款或贷款金额）。
- $ r $ 是年利率（以小数形式表示）。
- $ n $ 是每年复利的次数。
- $ t $ 是资金投资或借贷的年数。

在实际应用中，银行和金融机构会提供具体的年利率信息，包括是否为复利以及复利的频率等。

**简单年利率的例子**

假设你投资了1000元，年利率是5%。

计算步骤：
1. 计算年度利息：$ \text{年度利息} = \text{本金} \times \text{年利率} $
2. 计算公式：$ \text{年度利息} = 1000 \times 0.05 = 50 $元

结果：
- 一年后你将获得50元的利息。

**复利年利率的例子**

假设你投资了1000元，年利率是5%，并且利息每年复利一次。

计算步骤：
1. 计算一年后的金额：$ A = P \times (1 + r) $
2. 计算公式：$ A = 1000 \times (1 + 0.05) = 1000 \times 1.05 = 1050 $元

结果：
- 一年后你的投资总额将是1050元。

现在，让我们考虑一个更复杂的情况，如果利息每年复利两次。

**复利年利率（每年复利两次）的例子**

假设你投资了1000元，年利率是5%，并且利息每半年复利一次。

$计算步骤$：
1. 计算半年利率：$ \text{半年利率} = \frac{5\%}{2} = 2.5\% $
2. 计算半年后的金额：$ A_{\text{半年}} = P \times (1 + \frac{r}{n}) $
3. 计算公式：$ A_{\text{半年}} = 1000 \times (1 + 0.025) = 1025 $元
4. 再计算下一个半年后的金额：$ A_{\text{一年}} = A_{\text{半年}} \times (1 + \frac{r}{n}) $
5. 计算公式：$ A_{\text{一年}} = 1025 \times (1 + 0.025) = 1050.625 $元

结果：
- 一年后你的投资总额将是1050.625元。

在复利的情况下，由于利息在周期内被加入本金并产生新的利息，因此最终的金额会略高于简单年利率的计算结果。

**应收账款**（Accounts Receivable，AR）：截至到观察时间点，用户当前所在账期的应收账款。

**账龄**（Month of Book, MOB）：资产放款月份，MOB0就是第0个月，也就是没有满30天，MOB1就是放款后第一个完整的月份。

**逾期天数**（Days Past Due, DPD）：已逾契约书约定缴款日期的延滞天数，贷放型产品自到期当天开始计算，如DPD0是到期当日，DPD1是到期过一日。

**逾期期数**（Bucket）：逾期的月份数。逾期1月标记为M1，逾期2个月为M2。

**逾期阶段**（Stage）：分为前期，中期，后期和转呆账，一般将M1为前期，M2~M3为中期，M4列为后期，已转呆账者则列入呆账。

**即期指标**（Lagged）：计算延滞率时常用的两种方法之一，以当期逾期期数对应的滞纳金额/应收账款（AR）。

**递延指标**（Lagged）：计算延率时常用的两种方法之一，为延滞金额/上月应收账款。若单纯想了解各月资产质量结构，可使用即期指标，但若想精准追溯到逾放源头，建议采用递延指标。

**留存率**（RetainedRate）：实际分为人头留存率(用户复贷占比)和余额留存率(复贷金额占比)。提现率(Withdrawal Rate):使用提现功能的客户占比。

**额度使用率**（Credit Utilization Rate）：用户使用额度占总额度的百分比。

**复借率**（Reloan Rate）**用户还款后再次贷款的概率。主要用于循环贷产品。

# 信贷风险指标

**延滞率 (Delinquent Rate)**: 
  - 计算可分为即期和递延方式，观察特定 Bucket 以上的延滞率，如 M2+ 的 Lagged 和 M4+ 的 Lagged。
  - 例如 M2+ 的 Lagged，分母为两个月前应收账款，分子为本月 M2（含以上）尚未转呆账的逾期金额。
  - M1 落入 M2 以上可确认为无意还款或蓄意拖欠。

**不良率 (Bad Rate)**:
  - 当月不良资产数 / 总资产数。
  - 包含逾期户、债务协议及高风险控管户等。

**转呆账率 (Write-off，WO)**:
  - 当月转呆账金额 / 逾期开始月的应收账款。
  - 经过年化换算后，月转呆账率转换为年损失率。

**涂损失率 (Net Charge-off, NCL)**:
  - 当期转呆账金额 - 当期账款回收金额。
  - 计算公式：净损金额 / 逾期开始月的应收账款，常以年化形态展示。

**累计转呆账率**:
  - 计算期满客户的累计损失率。
  - 公式：分母为案件第1~(K+N)期的转呆账总金额，分子为已(K+N)期案件的初贷总金额。
  - N 表示转呆账所需期数，K 表示总期数。

**负债比 (Debt Burden Ratio, DBR)**:
  - 公式：总无担保债务归户后的总余额（信用卡、现金卡及信用贷款）/ 月收入。
  - 一般不超过 22 倍。

**月负比**:
  - 衡量还款压力的指标，计算公式：推估每月各项贷款月付额 + 最低生活费 / 月收入。

**平均额度**:
  - 用于观察不同产品及群组间额度的差异。

**风险等级 (Risk Grade)**:
  - 用于客户分群的指标，常通过信用评分划分。

**命中率 (Hit Rate)**:
  - 控管后一定时间内客户发生延滞的几率。
  - 用于信用卡的中途授信及早期预警，命中率过低可能表示风险判断有误。

**可用余额 (Open To Buy, OTB)**:
  - 观察控管命中客户时的信用卡可用余额，用于评估控管所减少的损失。

**迁徙率 (Flow Rate)**:
  - 观察前期逾期金额经过催收后，继续落入下一期的几率。

**首次还款逾期 (First Payment Delinquency, FPD)**:
  - FPD7 描述的是客户在最后还款日后7天内未还款且未延期的比例。
  - 其他常用指标包括 FPD30。

**预期损失 (Expected Loss, EL)**:
  - 根据历史数据预估策略或模型变动后的损失。

**收入负载 (Debt To Income, DTI)**:
  - 公式：每月应偿还债务 / 每月税前收入。

# 企业信贷风控架构

需要多人参与配合，通过多环节把控，才能够有效的控制风险和最大化收益。

信贷风控架构：



用户四要素包括：身份证号，姓名，手机号，银行卡号。平台设置四要素验证，根据申请用户填写的信息，同时联系相关机构校验用户是否为信息所有人。平台根据信用评分可以有效估计申请人的信用，并据此使用差异化定价手段为用户授信不同的额度。
