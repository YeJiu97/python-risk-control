## 基础概念

**探索性数据分析（Exploratory Data Analysis，EDA）**是数据分析过程中的一个关键步骤，旨在通过**统计图表和摘要统计量**，初步探索数据的结构、特征和潜在关系。EDA 是在数据建模之前的必要步骤，帮助分析人员深入理解数据集，发现异常、模式、趋势以及潜在的假设。

EDA 的主要目标：
1. **了解数据的基本结构**：通过可视化和统计方法，了解数据的分布、趋势、极值和异常值等基本信息。
2. **检测异常值和数据错误**：识别数据中的异常点和错误，确保数据质量。
3. **初步识别数据中的模式和关系**：通过分析特征间的相互关系，发现潜在的相关性或因果关系。
4. **指导后续的建模过程**：EDA 帮助选择合适的模型和算法，提供对建模假设的初步检验。

常用的 EDA 技术：
1. **数据分布的可视化**：
   - **直方图（Histogram）**：显示单一变量的分布情况。
   - **箱线图（Box Plot）**：揭示数据的中位数、四分位数范围及异常值。
   - **密度图（Density Plot）**：平滑的分布曲线，显示数据的概率密度。
2. **变量之间的关系**：
   - **散点图（Scatter Plot）**：显示两个变量之间的关系（特别适合数值型数据）。
   - **相关矩阵（Correlation Matrix）**：显示多个变量之间的线性相关性。
   - **热力图（Heatmap）**：通过颜色表示变量之间的相关性。
3. **数据的统计特性**：
   - **描述性统计（Descriptive Statistics）**：包括均值、方差、标准差、最小值、最大值、中位数、分位数等。
   - **频率表（Frequency Table）**：用于分类变量的频率分布。
4. **多变量关系分析**：
   - **成对关系图（Pairplot）**：可同时显示多个变量之间的关系。
   - **主成分分析（PCA）**：通过降维，展示数据的主要变化方向。

EDA 的步骤：
1. **数据加载和检查**：加载数据，检查数据的结构、大小、数据类型，以及是否存在缺失值。
2. **数据分布的初步分析**：查看每个变量的分布，了解数值型数据的范围、分布形态，以及分类变量的类别分布。
3. **变量之间的关系分析**：通过可视化工具分析变量之间的关联性，识别潜在的模式和趋势。
4. **处理异常值和缺失值**：发现并处理异常数据点或缺失值，以确保数据质量。
5. **得出初步结论**：根据分析结果，形成对数据的初步理解，并为后续建模打下基础。

EDA 的重要性：
- **理解数据特征**：EDA 帮助分析人员发现数据中的重要特征、趋势和潜在问题。
- **指导建模**：通过 EDA，分析人员可以选择合适的建模方法，避免盲目套用算法。
- **预防错误假设**：EDA 能识别不符合预期的数据特征，防止模型基于错误假设构建。

总结来说，探索性数据分析是数据分析的基础阶段，通过可视化和统计分析，帮助深入理解数据，为后续的建模与决策提供依据。

## 连续型数据

在进行**连续型变量**的探索性数据分析（EDA）时，了解和计算一些关键的统计指标可以帮助你全面了解数据的分布、集中趋势、离散程度以及异常点。以下是常见的**统计指标**，它们有助于全面描述连续型数据的特性：

1. **集中趋势（Central Tendency）**
   - **均值（Mean）**：数据集的平均值，表示所有数值之和除以数据点的个数。用于衡量数据的集中位置。
     - 计算公式：$\text{Mean} = \frac{1}{n}\sum_{i=1}^{n} x_i$
   - **中位数（Median）**：将数据排序后位于中间的值。如果数据点为偶数，则取中间两个数的平均值。中位数不受极端值影响。
   - **众数（Mode）**：数据集中出现频率最高的值，主要用于分析有明显频率特征的连续数据。
2. **离散程度（Dispersion）**
   - **方差（Variance）**：表示数据分布的离散程度。它衡量每个数据点与均值之间的平方距离的平均值。值越大，数据的波动越大。
     - 计算公式：$\text{Variance} = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2$
   - **标准差（Standard Deviation）**：方差的平方根，衡量数据分布的平均偏离程度。与方差相比，标准差与数据单位相同，更便于解释。
     - 计算公式：$\text{Standard Deviation} = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2}$
   - **极差（Range）**：最大值与最小值的差值，表示数据的总体波动范围。
     - 计算公式：$\text{Range} = \text{Max} - \text{Min}$
3. **位置指标（Positional Indicators）**
   - **四分位数（Quartiles）**：将数据按25%的间隔划分为四部分，主要用于描述数据的分布特性。
     - **第一四分位数（Q1）**：25%的数据低于此值。
     - **中位数（Q2）**：50%的数据低于此值。
     - **第三四分位数（Q3）**：75%的数据低于此值。
   - **四分位距（Interquartile Range, IQR）**：Q3 和 Q1 之间的差距，表示数据中间50%的分布范围，常用于检测异常值。
     - 计算公式：$\text{IQR} = Q3 - Q1$
4. **形状指标（Shape Indicators）**
   - **偏度（Skewness）**：衡量数据分布的对称性。如果偏度为0，表示分布对称；如果偏度大于0，表示右偏（正偏），即数据集中在左侧，右侧尾部较长；如果偏度小于0，表示左偏（负偏）。
     - 公式：$\text{Skewness} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{x_i - \bar{x}}{s} \right)^3$
   - **峰度（Kurtosis）**：衡量分布的尖峰程度。较高的峰度值表示数据在均值附近集中，较低的峰度值表示数据分布较为平坦。
     - 公式：$\text{Kurtosis} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{x_i - \bar{x}}{s} \right)^4 - 3$
5. **分布信息**
   - **最小值（Minimum）**：数据集中最小的值，显示数据分布的下界。
   - **最大值（Maximum）**：数据集中最大的值，显示数据分布的上界。
   - **百分位数（Percentiles）**：将数据分成100个等份，常用的有25%、50%、75%百分位数，分别对应Q1、中位数和Q3。可以用来描述不同部分的数据分布。
6. **异常值检测**
   - **异常值（Outliers）**：通常是指数据中偏离正常范围的值。通过计算四分位距 (IQR) 来检测异常值。若某个数据点小于 $ Q1 - 1.5 \times IQR $ 或大于 $ Q3 + 1.5 \times IQR $，则该点被认为是异常值。

**可视化方法：** 除了统计指标外，利用可视化工具可以更直观地进行探索性数据分析：
- **直方图（Histogram）**：用于查看数据的分布形态。
- **箱线图（Box Plot）**：可视化四分位数、IQR 和异常值。
- **密度图（Density Plot）**：显示数据的概率密度，帮助理解数据的分布形状。

总结来说，对于连续型变量的探索性数据分析，需要通过**集中趋势**、**离散程度**、**位置指标**和**形状指标**，结合可视化方法，全面了解数据的分布特征、潜在异常值和整体趋势。这些指标为后续的建模和决策提供了坚实的基础。

## 离散型数据变量

对于**离散型数据变量**的探索性数据分析（EDA），需要使用一些特定的统计指标和方法来描述数据的分布和特性。离散型数据通常是分类变量（如类别、标签）或整数变量（如计数数据）。以下是分析离散型数据时常用的统计指标和方法：

1. **频率分布（Frequency Distribution）**
   - **频数（Frequency）**：每个类别或离散值在数据集中出现的次数。
   - **频率（Relative Frequency）**：某一类别出现的频数占总数据的比例。计算方法是该类别的频数除以总样本数。
     - 计算公式：$\text{Relative Frequency} = \frac{\text{Frequency of a Category}}{\text{Total Frequency}}$
   - **百分比频率（Percentage Frequency）**：相对频率乘以100，表示为百分比。
2. **众数（Mode）**
   - **众数（Mode）**：离散型数据中出现次数最多的值，特别适合于分类数据。众数反映了数据集中的主流类别。
3. **分类变量的分布情况**
   - **类别数量（Number of Categories）**：了解离散变量的所有可能值（类别）的数量。对于分类变量，类别的多少可以帮助判断变量的稀疏程度。
   - **类别均匀性（Category Uniformity）**：通过比较每个类别的频数，判断类别之间的分布是否均匀，是否有某些类别特别占优或罕见。
4. **交叉表（Crosstab/Contingency Table）**
   - **交叉表（Crosstab）**：用于分析两个或多个离散变量之间的关系。交叉表展示了变量的不同类别之间的联合分布情况，帮助理解它们之间是否存在关联。
   - **列联表（Contingency Table）**：类似于交叉表，通常用于检验两个分类变量之间的独立性。可以结合卡方检验（Chi-Square Test）进行进一步的分析。
5. **集中趋势和离散程度（仅适用于有序离散变量）**
   - 对于**有序离散变量**（例如评分等级：1、2、3、4、5），虽然是离散型数据，但也可以使用一些类似于连续变量的统计指标。
     - **中位数（Median）**：将数据排序后，找到位于中间的那个值。
     - **百分位数（Percentiles）**：例如第25、第50（即中位数）、第75百分位数，可以显示有序离散数据的分布情况。
6. **卡方检验（Chi-Square Test）**
   - 用于检测两个分类变量之间是否存在显著的相关性或独立性。通过比较实际观察到的频数与期望频数之间的差异，判断变量之间是否有统计上的关联。
   - **卡方检验统计量**：$\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}$，其中 $O_i$ 为观测频数，$E_i$ 为期望频数。
7. **离散型数据的描述统计**
   - 对于一些离散型数据（如计数数据），也可以计算平均值、方差等描述性统计量，虽然这些指标通常用于连续型数据，但在分析某些离散型数据时同样适用。
     - **均值（Mean）**：虽然离散数据不能形成真正的“均值”，但可以通过计数数据的总和除以样本数量来衡量平均情况。
     - **方差和标准差（Variance & Standard Deviation）**：用于衡量离散型数据的波动范围，特别是对计数数据，能够评估其离散程度。
8. **熵（Entropy）**
   - **熵（Entropy）**：用于衡量分类变量的分布信息量。如果熵值高，说明类别分布较为均匀；熵值低，则说明某些类别占主导地位。
     - 计算公式：$\text{Entropy} = - \sum p(x) \log p(x)$，其中 $p(x)$ 为类别的相对频率。
9. **Gini 系数（Gini Index）**
   - **Gini 系数**是一种用于衡量分类数据中不纯度的指标，尤其在决策树算法中常用。Gini 系数的值越小，说明数据的纯度越高。
     - 计算公式：$\text{Gini Index} = 1 - \sum p(x)^2$，其中 $p(x)$ 为类别的相对频率。
10. **可视化方法**
   - **条形图（Bar Plot）**：展示分类变量的频数或频率分布。条形图是离散型数据最常用的可视化工具之一。
   - **饼图（Pie Chart）**：显示分类变量的比例分布，适合用来展示类别占比。
   - **马赛克图（Mosaic Plot）**：用于展示多个分类变量之间的关系。
   - **分布条形图（Stacked Bar Plot）**：用于展示离散型数据的分布以及不同类别之间的关系，适合分析多变量之间的相互作用。

**总结：**
离散型数据的探索性分析主要集中在**频率分布**和**类别关系**的分析上。需要通过**频数、百分比、交叉表**等工具来分析离散变量的分布情况，并使用**熵、Gini 系数**等指标衡量数据的不纯度或类别的分布均匀性。常用的可视化工具如条形图和交叉表，能帮助清晰地展示离散型数据的结构和关系。

这些指标帮助你从多个维度分析离散型数据，特别是分类和计数数据的特点、分布和相关性，为后续的建模和决策提供依据。

## 缺失值处理

**缺失值处理**是数据清洗中的重要步骤，尤其在探索性数据分析（EDA）或建模之前，处理不当的缺失值会对分析结果和模型性能产生负面影响。处理缺失值的方法根据数据集的特性和缺失的模式而有所不同。以下是常见的缺失值处理方法和其适用的情况：

1. **确定缺失值的类型和模式**
   在处理缺失值之前，首先需要了解缺失值的**类型**和**模式**，这有助于选择合适的处理方法：
   - **完全随机缺失（MCAR）**：缺失值的出现与任何观察到的或未观察到的变量无关。
   - **随机缺失（MAR）**：缺失值的出现与其他观测变量有关，但与缺失的变量本身无关。
   - **非随机缺失（MNAR）**：缺失值的出现与缺失的变量本身有关。
   - 了解缺失值的模式可以通过统计方法（如 Little's MCAR Test）或可视化方法（如缺失值矩阵图）来实现。
2. **删除缺失值**
   - **删除含有缺失值的行（Row-wise Deletion or Listwise Deletion）**
     - 适用情况：如果缺失值占比很小（通常低于5%）且随机分布，直接删除含有缺失值的行不会显著影响结果。
     - 优点：简单直接，不引入任何假设。
     - 缺点：如果缺失值较多，会导致大量数据丢失，可能会引入偏差。
   - **删除含有缺失值的列（Column-wise Deletion）**
     - 适用情况：当某一列的缺失值非常多（如50%以上），该列可能对模型没有帮助，可以直接删除。
     - 优点：保留数据集中有用的信息。
     - 缺点：可能丢失重要的变量，尤其是当列有显著解释能力时。
3. **替代缺失值（Imputation）**
   - **均值替代（Mean Imputation）**
     - 适用情况：用于连续型变量，缺失值随机且数据呈正态分布。
     - 方法：用该列非缺失值的均值来填补缺失值。
     - 缺点：会低估数据的方差，无法反映数据的真实波动性，容易引入偏差。
   - **中位数替代（Median Imputation）**
     - 适用情况：适合于存在离群值的连续型变量。
     - 方法：用该列的中位数替换缺失值，较少受极端值的影响。
     - 优点：在数据存在偏斜的情况下比均值替代效果更好。
   - **众数替代（Mode Imputation）**
     - 适用情况：适合离散型或分类变量。
     - 方法：用该列出现频率最高的值填补缺失值。
     - 优点：简单有效，不会改变分类变量的分布。
   - **K近邻替代（K-Nearest Neighbors Imputation, KNN）**
     - 适用情况：适合连续型和分类型变量，缺失值不是随机分布的情况下。
     - 方法：利用与缺失值行最相似的K个最近邻的观测数据来填补缺失值，计算基于欧几里得距离或其他距离度量。
     - 优点：保持数据之间的相关性，填补效果好。
     - 缺点：计算复杂，数据量大时处理速度慢。
   - **回归插补（Regression Imputation）**
     - 适用情况：适合连续型变量，缺失值与其他变量高度相关的情况下。
     - 方法：使用其他变量作为自变量，通过回归模型预测缺失值。
     - 优点：考虑了其他变量间的关系，能有效预测缺失值。
     - 缺点：容易引入过拟合风险，且不增加数据的方差。
   - **多重插补（Multiple Imputation）**
     - 适用情况：适合各种变量类型，尤其当缺失值较多时。
     - 方法：多次生成缺失值的不同填补值，得到多个完整数据集，之后通过多次模拟得到最终结果。
     - 优点：考虑了数据的不确定性，减少了插补偏差，保留了数据的变异性。
     - 缺点：较为复杂，计算开销大。
4. **插值法（Interpolation）**
   - **线性插值（Linear Interpolation）**
     - 适用情况：适用于时间序列数据。
     - 方法：根据相邻已知值通过线性方式估计缺失值。
     - 优点：适用于缺失值较少、数据较平滑的情况。
     - 缺点：无法处理非线性趋势的变化，插值精度有限。
   - **样条插值（Spline Interpolation）**
     - 适用情况：适用于时间序列或连续型数据，且数据有较复杂的变化趋势。
     - 方法：使用多项式函数拟合数据，用拟合的曲线填补缺失值。
     - 优点：能较好地适应数据的变化趋势，精度较高。
     - 缺点：计算复杂度较高。
5. **预测模型插补**
   - 使用机器学习模型（如决策树、随机森林）来预测缺失值。这种方法通过构建模型预测缺失值较为精准。
     - **随机森林插补（Random Forest Imputation）**：基于随机森林算法构建模型填补缺失值。通过训练模型预测缺失数据点的值。
     - **优点**：考虑了多个变量之间的复杂关系，适合大规模数据集。
     - **缺点**：需要较大的计算资源，特别是对于高维数据。
6. **标记缺失值**
   - **新增缺失值指示变量（Missing Indicator）**
     - 适用情况：在某些情况下，缺失值本身可能包含某种信息。可以创建一个新的二进制变量来表示某一列是否存在缺失值。
     - 方法：添加一个新列，标记每个数据点是否缺失。
     - 优点：保留了缺失值的信息，避免信息丢失。
     - 缺点：适用于少量的缺失数据，但如果缺失值太多，会增加模型复杂性。
7. **不处理缺失值**
   - 在某些建模算法中（如决策树、XGBoost等），模型可以自动处理缺失值，尤其当数据量大且缺失值不多时，可以选择让模型直接处理。
   - 优点：不需要手动填补，避免可能引入的假设偏差。
   - 缺点：缺失值太多时，可能影响模型性能。

**总结**

缺失值的处理方式应根据数据的特性和缺失值的模式来选择。对于少量随机缺失值，可以通过删除或简单插补（均值、中位数等）来处理；而对于大量或非随机的缺失值，采用高级的插补方法（如KNN、多重插补、回归插补）会更加合适。对于时间序列数据，插值法也是一种常用的解决方案。

## 特征生成

特征生成是数据预处理和机器学习中的一个重要步骤，涉及从原始数据中创建新特征，以提高模型的性能。良好的特征能够帮助模型更好地理解数据的结构和关系，从而提升预测准确性。特征生成通常包括新特征的创建、特征选择与降维、数据清洗与转换、外部数据集的整合，以及特征的时间序列处理。

在特征生成中，新特征的创建可能包括组合特征，例如将“收入”和“支出”两个特征结合生成“可支配收入”特征，或对现有特征进行数学变换，如取对数、平方、平方根等，以减少数据的偏态分布。分类特征编码也是常见的技术，通常使用独热编码或标签编码将分类特征转换为数值特征。此外，从时间戳中提取有用的信息，如年份、月份、星期几和小时等，都是有效的特征生成方法。

在生成大量特征后，特征选择和降维变得至关重要。使用特征选择技术（如Lasso回归、决策树的重要性）来筛选出对模型有显著影响的特征可以提升模型性能。降维技术，如主成分分析（PCA）和线性判别分析（LDA），则可以将高维特征空间降维，以简化模型并减少计算复杂度。

在特征生成过程中，数据清洗与转换也是不可或缺的一环。处理缺失值、标准化和归一化数值特征，确保生成的特征是完整的且在相同的尺度上，都是重要的步骤。此外，将外部数据源（如市场数据、人口统计数据等）引入到现有数据集中，以丰富特征集，也能提供更多上下文信息。

在处理时间序列数据时，可以生成滞后特征、滚动统计特征等，以捕捉时间依赖性。这些方法有助于在时间序列分析中更准确地捕捉数据趋势。特征生成的主要目标是提升模型性能，通过创建能更好地表达数据特性的特征，帮助模型提高预测准确性，同时减少过拟合风险，确保模型的泛化能力。总体而言，特征生成是一个创造性和迭代性的过程，结合领域知识和经验往往能够取得更好的效果。

```python
import pandas as pd
import numpy as np

# 创建一个示例数据集，包含一个日期列
data = {
    'date': pd.date_range(start='2022-01-01', periods=10, freq='D'),
    'sales': [150, 200, 300, 400, 250, 500, 600, 700, 800, 900],
}

df = pd.DataFrame(data)

# 生成新特征：提取星期几和是否为周末
df['weekday'] = df['date'].dt.weekday  # 0 = Monday, 6 = Sunday
df['is_weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)  # 1 if Saturday/Sunday, 0 otherwise

# 生成额外特征：提取年份、月份和日期
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

# 输出处理后的数据集
print(df)
```

## 特征聚合

特征聚合是一种数据预处理技术，涉及将多个相关特征结合成一个更高层次的特征，以简化数据集并减少维度。这种方法通常用于提高模型的性能和解释性，同时帮助减少过拟合的风险。特征聚合可以通过不同的方式进行，常见的包括求和、平均、计数、最大值、最小值等统计方法。

特征聚合的关键目标是提取重要的信息，使模型更容易捕捉数据中的趋势和模式。例如，在处理时间序列数据时，可以对特定时间段内的销售数据进行聚合，以生成新的特征，如每周或每月的总销售额。这种方法不仅可以减少数据的复杂性，还能提高模型对长期趋势的识别能力。

特征聚合的应用场景包括：

1. **时间序列数据**：在时间序列分析中，可以对每日数据进行聚合，生成周、月或季度的统计数据，例如每周的平均销售额或每月的总销量。

2. **用户行为数据**：在电子商务平台上，可以对用户的购买行为进行聚合，计算出每个用户在过去一个月内的总消费额、购买次数等。

3. **地理数据**：在处理地理信息时，可以将不同地区的特征进行聚合，例如计算某一城市的总人口、平均收入等。

在进行特征聚合时，需要注意以下几点：

- 聚合特征的选择：应选择与目标变量高度相关的特征进行聚合，以确保信息的有效性。
- 处理缺失值：在聚合过程中，需要妥善处理缺失值，以避免对聚合结果产生负面影响。
- 过度聚合的风险：过度聚合可能会导致信息的丢失，因此在设计聚合特征时，需要平衡特征的数量与信息量。

总的来说，特征聚合是数据分析和机器学习中的重要工具，能够帮助数据科学家有效地提取和利用数据中的有用信息。通过合理的聚合，可以提高模型的性能和可解释性，从而在实际应用中取得更好的结果。

```python
import pandas as pd
import numpy as np

# 创建一个示例数据集
data = {
    'user_id': ['user_1', 'user_1', 'user_1', 'user_2', 'user_2', 'user_2', 'user_3', 'user_3'],
    'date': pd.date_range(start='2022-01-01', periods=8, freq='D'),
    'sales': [150, 200, 300, 400, 250, 500, 600, 700],
}

df = pd.DataFrame(data)

# 打印原始数据集
print("原始数据集：")
print(df)

# 将日期列转换为月份
df['month'] = df['date'].dt.to_period('M')

# 进行特征聚合：计算每个用户每月的总销售额和平均销售额
aggregated_df = df.groupby(['user_id', 'month']).agg(
    total_sales=('sales', 'sum'),
    average_sales=('sales', 'mean')
).reset_index()

# 打印聚合后的数据集
print("\n聚合后的数据集：")
print(aggregated_df)
```

## 特征组合

特征组合是数据预处理中的一种技术，涉及将两个或多个现有特征结合在一起，以创建新的特征。通过特征组合，数据科学家可以生成更复杂的特征，从而帮助模型更好地捕捉数据中的潜在模式和关系。这种方法在机器学习中非常重要，尤其是在处理非线性关系和特定类型的数据时。

特征组合可以通过多种方式实现，常见的方法包括：

1. **数学运算**：将两个或多个特征进行加、减、乘、除等运算。例如，可以将房屋面积与价格结合，以生成每平方英尺的价格特征。

2. **分类特征的组合**：对分类特征进行组合，以生成新的分类特征。例如，可以将“性别”和“教育水平”两个特征结合，创建一个新的特征“性别与教育组合”。

3. **多项式特征**：通过生成特征的幂次或交互项来组合特征，例如生成`x1^2`、`x2^2`、`x1 * x2`等多项式特征，以捕捉特征之间的非线性关系。

特征组合的主要优点在于能够增强模型的表达能力，使模型能够更好地拟合复杂的关系。然而，特征组合也可能带来一些挑战，比如：

- **维度灾难**：随着特征数量的增加，组合的特征数量也会大幅增加，导致计算复杂性上升和模型训练变得更加困难。

- **过拟合**：过多的特征组合可能导致模型在训练数据上表现良好，但在测试数据上性能较差。

在实践中，特征组合是一种重要的特征工程方法，适用于许多机器学习任务。合理选择和创造特征组合可以提高模型的性能和准确性。

```python
import pandas as pd

# 创建一个示例数据集，包含房屋特征
data = {
    'house_id': [1, 2, 3, 4],
    'bedrooms': [3, 2, 4, 3],
    'bathrooms': [2, 1, 3, 2],
    'area_sqft': [1500, 800, 2000, 1200],
    'price': [300000, 180000, 500000, 350000],
}

df = pd.DataFrame(data)

# 打印原始数据集
print("原始数据集：")
print(df)

# 特征组合：计算总房间数（卧室 + 浴室）
df['total_rooms'] = df['bedrooms'] + df['bathrooms']

# 特征组合：计算每平方英尺的价格
df['price_per_sqft'] = df['price'] / df['area_sqft']

# 打印组合后的数据集
print("\n组合后的数据集：")
print(df)
```

## 特征变换

WOE（Weight of Evidence，证据权重）是一种用于处理分类变量的技术，尤其在信用评分和风险管理中应用广泛。WOE通过将每个类别的事件率（如好客户和坏客户的比例）与整体事件率进行比较，以帮助分析和解释变量与目标变量之间的关系。

WOE的主要思想是将每个类别的坏客户比例（坏客户/总客户）与好客户比例（好客户/总客户）进行比较，从而生成一个相对权重。这种方法能够有效地减少分类变量的数量，提高模型的稳定性，并降低过拟合的风险。

WOE的计算公式如下：

$
\text{WOE} = \ln\left(\frac{P(\text{Good})}{P(\text{Bad})}\right)
$

其中，$P(\text{Good})$ 是某个类别的好客户比例，$P(\text{Bad})$ 是该类别的坏客户比例。

WOE的特点包括：

1. **单调性**：WOE的值在一定程度上反映了变量与目标变量之间的单调关系，通常可以与线性模型结合使用。

2. **可解释性**：WOE提供了一个直观的方式来理解分类变量的影响，便于业务人员进行解释。

3. **处理缺失值**：在WOE计算中，可以将缺失值单独处理，从而避免对模型的影响。

4. **提升模型性能**：使用WOE转换后的变量通常能够提高模型的预测能力，尤其是在信用评分和风险分析中。

总的来说，WOE是信用评分和其他领域中常用的数据预处理技术，能够将分类变量转化为数值型特征，为后续建模提供有力支持。

```python
import pandas as pd
import numpy as np

# 创建示例数据集
data = {
    'customer_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'default': [0, 0, 1, 1, 0, 0, 1, 1],  # 0 表示好客户，1 表示坏客户
    'age_group': ['18-25', '18-25', '26-35', '26-35', '36-45', '36-45', '46-55', '46-55']
}

df = pd.DataFrame(data)

# 打印原始数据集
print("原始数据集：")
print(df)

# 计算每个年龄组的坏客户和好客户的数量
woe_df = df.groupby('age_group').agg(
    bad_count=('default', 'sum'),
    good_count=('default', lambda x: len(x) - sum(x))
).reset_index()

# 计算总坏客户和总好客户的数量
total_bad = woe_df['bad_count'].sum()
total_good = woe_df['good_count'].sum()

# 计算每个年龄组的坏客户比例和好客户比例
woe_df['bad_rate'] = woe_df['bad_count'] / total_bad
woe_df['good_rate'] = woe_df['good_count'] / total_good

# 计算WOE，使用np.where来处理除以零的情况
woe_df['WOE'] = np.where(
    (woe_df['good_rate'] == 0) & (woe_df['bad_count'] > 0),
    -np.inf,  # 如果没有好客户，设置WOE为负无穷
    np.where(
        (woe_df['bad_rate'] == 0) & (woe_df['good_count'] > 0),
        np.inf,  # 如果没有坏客户，设置WOE为正无穷
        np.log(woe_df['good_rate'] / woe_df['bad_rate'])
    )
)

# 打印WOE结果
print("\nWOE结果：")
print(woe_df[['age_group', 'bad_count', 'good_count', 'WOE']])
```

```python
# 假设我们要将 WOE 值添加到原始数据集中
df = df.merge(woe_df[['age_group', 'WOE']], on='age_group', how='left')

# 打印合并后的数据集
print("\n合并后的数据集：")
print(df)

# 现在 df 中的 WOE 列可以用作特征输入到模型中
```

## 信息量

信息值（Information Value，IV）是一种用于评估变量（通常是分类变量）在区分好客户和坏客户（或二分类问题）中的重要性或预测能力的度量。信息值通常与权重（Weight of Evidence，WOE）结合使用，以帮助构建更有效的信用评分模型。

信息值的计算基于每个类别的好客户和坏客户的比例，能够反映某个变量对于区分这两类的贡献。IV 值越高，表示该变量对模型的预测能力越强。

信息值的计算公式如下：

$
IV = \sum (P_{bad} - P_{good}) \times WOE
$

其中：

- $P_{bad}$ 是在该类别中的坏客户比例。
- $P_{good}$ 是在该类别中的好客户比例。
- $WOE$ 是该类别的权重。

信息值的解读

- **IV < 0.02**: 变量的预测能力弱，不适合用于建模。
- **0.02 ≤ IV < 0.1**: 变量的预测能力一般，可以考虑使用。
- **0.1 ≤ IV < 0.3**: 变量的预测能力较强，值得用于建模。
- **IV ≥ 0.3**: 变量的预测能力非常强，强烈推荐用于建模。

## KS值

KS值（Kolmogorov-Smirnov值）是一种用于评估模型性能的指标，特别是在二分类问题中。它用于衡量两个分布之间的最大差异，通常是在预测模型的正类和负类的分布之间计算的。

KS值的计算方法：
1. **定义正负样本**：首先，将样本分为两类：正类（如违约客户）和负类（如良好客户）。
2. **计算累积分布函数（CDF）**：分别计算正类和负类样本的累积分布函数（CDF）。
3. **计算差异**：在每个可能的分值上，计算两个CDF之间的差异。
4. **找到最大差异**：KS值是这些差异中的最大值。

KS值的意义：
- **范围**：KS值的范围是从0到1，值越大表示模型的区分能力越强。
- **阈值**：
  - **KS值 ≤ 0.2**：模型的区分能力较弱。
  - **0.2 < KS值 ≤ 0.4**：模型的区分能力一般。
  - **KS值 > 0.4**：模型的区分能力较强。

应用：
KS值常用于信用评分模型、欺诈检测、风险管理等领域，以帮助评估模型在区分不同类型的样本（例如违约与非违约）方面的效果。通过比较模型的KS值，可以选择性能更优的模型。

```python
# Traditional Credit Scoring Using Logistic Regression
import scorecardpy as sc

# data prepare ------
# load germancredit data
dat = sc.germancredit()

# filter variable via missing rate, iv, identical value rate
dt_s = sc.var_filter(dat, y="creditability")

# breaking dt into train and test
train, test = sc.split_df(dt_s, 'creditability').values()

# woe binning ------
bins = sc.woebin(dt_s, y="creditability")
# sc.woebin_plot(bins)

# binning adjustment
# # adjust breaks interactively
# breaks_adj = sc.woebin_adj(dt_s, "creditability", bins) 
# # or specify breaks manually
breaks_adj = {
    'age.in.years': [26, 35, 40],
    'other.debtors.or.guarantors': ["none", "co-applicant%,%guarantor"]
}
bins_adj = sc.woebin(dt_s, y="creditability", breaks_list=breaks_adj)

# converting train and test into woe values
train_woe = sc.woebin_ply(train, bins_adj)
test_woe = sc.woebin_ply(test, bins_adj)

y_train = train_woe.loc[:,'creditability']
X_train = train_woe.loc[:,train_woe.columns != 'creditability']
y_test = test_woe.loc[:,'creditability']
X_test = test_woe.loc[:,train_woe.columns != 'creditability']

# logistic regression ------
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty='l1', C=0.9, solver='saga', n_jobs=-1)
lr.fit(X_train, y_train)
# lr.coef_
# lr.intercept_

# predicted proability
train_pred = lr.predict_proba(X_train)[:,1]
test_pred = lr.predict_proba(X_test)[:,1]

# performance ks & roc ------
train_perf = sc.perf_eva(y_train, train_pred, title = "train")
test_perf = sc.perf_eva(y_test, test_pred, title = "test")

# score ------
card = sc.scorecard(bins_adj, lr, X_train.columns)
# credit score
train_score = sc.scorecard_ply(train, card, print_step=0)
test_score = sc.scorecard_ply(test, card, print_step=0)

# psi
sc.perf_psi(
  score = {'train':train_score, 'test':test_score},
  label = {'train':y_train, 'test':y_test}
)
```
